```python
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx
% matplotlib inline
```

#### Identify and load a network dataset that has some categorical information available for each node.

```python
airport = pd.read_csv("https://raw.githubusercontent.com/saayedalam/Web-Analytics/master/Project1_data.csv")
airport.head()
```

```python
plt.figure(figsize=(16, 16))

# Create Directional graph
flight_graph = nx.from_pandas_edgelist(airport, source='Nodes_1', target='Nodes_2', create_using=nx.Graph())

print nx.info(flight_graph)

sp = nx.spring_layout(flight_graph)

n1 = list(airport.Nodes_1.unique())
n2 = list(airport.Nodes_2.unique())

# Source nodes
nx.draw_networkx_nodes(flight_graph, sp, nodelist=n1,aplha = 0.2, node_size=1000, node_color='blue')

# Draw target airports
nx.draw_networkx_nodes(flight_graph, sp, nodelist=n2, node_color='purple', node_size=100)

# Draw common Airports or nodes with more than 1 connection
pop_airport = [n2 for n2 in n2 if flight_graph.degree(n2) > 1]
nx.draw_networkx_nodes(flight_graph, sp, nodelist=pop_airport,node_color='green', node_shape = 'h', node_size=200)

nx.draw_networkx_edges(flight_graph, sp, width=1, edge_color="gray", alpha = 0.2)

node_labels = dict(zip(n1, n1))
nx.draw_networkx_labels(flight_graph, sp, labels=node_labels)

plt.axis('off')

plt.title("Airport")

plt.show()
```

#### For each of the nodes in the dataset, calculate degree centrality and eigenvector centrality.

Degree centrality is the measure of how many neighbors a node has. With this advantage, nodes that have more edges may have multiple ways of accessing resources to reach goals.

Eigenvector centrality measures a node’s importance based on the importance of its neighbors.  Unlike degree centrality, it does not take into account how many connection a node has. That is, a person with few connections may have a higher eigenvector centrality measure depending on how well connected those connections are. 

```python
#nx.degree_centrality(flight_graph)

tbl_centrality = pd.DataFrame()
tbl_centrality['degree_centrality'] = nx.degree_centrality(flight_graph).values()
tbl_centrality['eigenvector_centrality'] = nx.eigenvector_centrality(flight_graph).values()
tbl_centrality.index = list(flight_graph)
tbl_centrality
```
